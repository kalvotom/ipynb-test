Pro přirozené $k$ definujme množiny (`align` prostředí)

\begin{align*}
  A_k &= \langle k, +\infty), &
  B_k &= \{ p \mid p \ \text{je prvočíslo a} \ p < k \}
\end{align*}

a (obyčejná samostatná rovnice)

$$
  C_k = (-k, k).
$$
